```python
from gensim.models import Word2Vec


sentences =[ ['Hello', 'I', 'am', 'learning', 'Natural', 'Language', 'Processing', 'with', 'Gensim'],
            ['Hello', 'I', 'am', 'learning', 'Natural', 'Language', 'Processing', 'with', 'Gensim'],
            ['my', 'name', 'is', 'Het', 'and', 'I', 'am', 'learning', 'Natural', 'Language', 'Processing', 'with', 'Gensim']
]


print(sentences)
model = Word2Vec(sentences,vector_size=5, window=3, min_count=1,)
```

```python
print("\n1. Finding Similarity\n")

print(model.wv.most_similar("learning"))
print(model.wv.most_similar("Gensim"))
print(model.wv.most_similar("Het"))
```

```python
print("\n2. Looking at Vectors\n")

learning = model.wv['learning']
print(learning[:5])
```

```python
print("\n3. Word Math")

try:
    result = model.mv.most_similiar(
        positive=['learning', 'Gensim'],
        negative=['am'],
        topn=1
    )
    print(f"learning + Gensim - am = {result[0][0]}")
except:
    print("Error")
```
