```python
import nltk
from nltk.corpus import brown
nltk.download('brown')
```

```python
brown_tag_sen = brown.tagged_sents(categories="adventure")

print(brown_tag_sen[0])
```

```python
brown.categories()
```

```python
size = int(len(brown_tag_sen)*0.9)
train_sents = brown_tag_sen[:size]
pos_tagger = nltk.HiddenMarkovModelTagger.train(train_sents)


def pos_tag(sentence):
    return pos_tagger.tag(nltk.word_tokenize(sentence))


print(pos_tag("The quick brown fox jumps over the lazy dog."))
```
