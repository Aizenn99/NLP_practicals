```python


text = "Hello, world world ! This is a test sentence."

words = text.split()  
```

```python
print(words)
```

```python
words_count = len(words)
print("Number of words:", words_count)
```

```python
clean_text = {word for word in text if word.isalpha()}
print("Clean words (alphabetic only):", clean_text)
```

```python
unique_words = set(words)
print(type(unique_words))
print("Unique words:", unique_words)
```

```python
#average word count 
total_word_length = sum(len(word)  for word in words)

average_word_length = total_word_length / words_count
print("Average word length:", average_word_length)
```

```python
#frequency of words
word_frequency = {}
for word in words:
    word = word.lower()  
    if word in word_frequency:
        word_frequency[word] += 1
    else:
        word_frequency[word] = 1

print("Word Frequency:", word_frequency)


```

```python
most_common_word = ""
max_count = 0
for word, count in word_frequency.items():
    if count > max_count:
        most_common_word = word
        max_count = count

print("Most common word:", most_common_word)
```

```python
def text_analysis(text):
    words = text.split()
    words_count = len(words)

    if words_count == 0:
        print("No words entered.")
        return

    clean_words = [word for word in text if word.isalpha()]
    unique_words = set(words)

    total_word_length = sum(len(word) for word in clean_words)
    average_word_length = total_word_length / words_count

    word_frequency = {}
    for word in clean_words:
        word = word.lower()
        if word in word_frequency:
            word_frequency[word] += 1
        else:
            word_frequency[word] = 1

    most_common_word = ""
    max_count = 0
    for word, count in word_frequency.items():
        if count > max_count:
            most_common_word = word
            max_count = count        
    
    print("Number of words:", words_count)
    print("Clean words (alphabetic only):", clean_words)
    print("Unique words:", unique_words)
    print("Average word length:", average_word_length)
    print("Word Frequency:", word_frequency)
    print("Most common word:", most_common_word)


input_text = input("Enter text for analysis: ")
text_analysis(input_text)

```
