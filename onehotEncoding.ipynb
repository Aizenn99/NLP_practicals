```python
txt = "Hello, I am learning Python   programming!"


split_txt = txt.split()
print(split_txt)

cleantxt = [word.strip("!.,?") for word in split_txt]
print(cleantxt)

char_list = list(cleantxt)
print(char_list)

```

```python
words_set = set(char_list)
print(words_set)
```

```python
sorted(words_set)
print(sorted(words_set))
```

```python
def one_hot_encoding(word, words_set):
    one_hot_vector = [0] * len(words_set)
    index = list(sorted(words_set)).index(word)
    one_hot_vector[index] = 1
    return one_hot_vector

for word in char_list:
    print(f"Word: {word} : {one_hot_encoding(word, words_set)}")
```
