```python

txt = "Hello, This is N-gram Text!"

split_txt = txt.split()
print(split_txt)
```

```python
clean_txt = [word.strip(".,!?:;") for word in split_txt]
print(clean_txt)
```

```python
for i in range(len(clean_txt)-1):
    bigram = (clean_txt[i],clean_txt[i+1],)
    print(bigram)
```

```python
for i in range(len(clean_txt)-2):
    trigram = (clean_txt[i],clean_txt[i+1],clean_txt[i+2])
    print(trigram)
```
